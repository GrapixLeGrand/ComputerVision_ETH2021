```python
import numpy as np
```

```python
points_W = np.linspace(8, 56, num=8, dtype=int)
points_H = np.linspace(8, 72, num=10, dtype=int)
print(points_W)
print(points_H)
xW, xH = np.meshgrid(points_W, points_H)

xW = xW.flatten()
xH = xH.flatten()
print(xW)
print(xH)
#xW = xW[:, :, None]
#xH = xH[:, :, None]
print(xW)
print(xH)
vPoints = np.stack((xW, xH), axis=1)
print(vPoints)
```
